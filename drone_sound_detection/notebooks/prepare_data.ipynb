```python
import os
import librosa
import soundfile as sf
from tqdm import tqdm

SOURCE_DRONES = "../data/drones"
SOURCE_NOT = "../data/not_drones"

TARGET_ROOT = "../clean"
TARGET_DRONES = os.path.join(TARGET_ROOT, "drones")
TARGET_NOT = os.path.join(TARGET_ROOT, "not_drones")

os.makedirs(TARGET_DRONES, exist_ok=True)
os.makedirs(TARGET_NOT, exist_ok=True)
```

```python
def collect_files(root):
    audio_files = []
    for r, _, fs in os.walk(root):
        for f in fs:
            if f.lower().endswith((".wav", ".mp3", ".m4a", ".ogg")):
                audio_files.append(os.path.join(r, f))
    return audio_files

print("сбор дронов...")
drone_files = collect_files(SOURCE_DRONES)
print("найдено:", len(drone_files))

print("сбор не_дронов...")
not_files = collect_files(SOURCE_NOT)
print("найдено:", len(not_files))
```

```python
def convert_and_save(file_list, target_dir, prefix):
    idx = 1
    for path in tqdm(file_list, desc=prefix):
        try:
            audio, sr = librosa.load(path, sr=None, mono=True)
        except Exception:
            continue

        new_name = f"{prefix}_{idx}.wav"
        new_path = os.path.join(target_dir, new_name)

        sf.write(new_path, audio, sr)
        idx += 1

convert_and_save(drone_files, TARGET_DRONES, "drone")
convert_and_save(not_files, TARGET_NOT, "not_drone")

print("готово")
```
